```python
df1 = df.pivot_table(index='user',columns='title',values='rating')
```

```python
df1.head()
```

```python
df1
```

```python
#Impute those NaNs with 0 values
df1.fillna(0, inplace=True)
```

```python
df1
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
user_sim = 1 -  pairwise_distances(df1.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
user_sim_df
```

```python
#Set the index and column names to user ids 
user_sim_df.index = df.user.unique()
user_sim_df.columns = df.user.unique()
```

```python
user_sim_df.iloc[0:15, 0:15]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:15, 0:15]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:15]
```
